```python
import random
import matplotlib.pyplot as plt

def simular_saltos(n, num_experimentos):
    total_saltos = 0
    promedios = []
    for i in range(1, num_experimentos + 1):
        a, b = random.sample(range(1, n+1), 2)
        saltos = abs(a - b)
        total_saltos += saltos
        promedio_actual = total_saltos / i
        promedios.append(promedio_actual)
    return promedios

def promedio_teorico(n):
    return (n + 1) / 3

nodos = [10, 100, 1000]
experimentos = 10000

resultados = {n: simular_saltos(n, experimentos) for n in nodos}

print("Resultados finales:")
for n in nodos:
    promedio_simulado = resultados[n][-1]
    print(f"Red con {n} nodos: Promedio de saltos = {promedio_simulado:.2f}")

fig, axs = plt.subplots(3, 1, figsize=(10, 15))
for i, n in enumerate(nodos):
    teoria = promedio_teorico(n)
    axs[i].plot(resultados[n], label='Simulación')
    axs[i].axhline(teoria, color='r', linestyle='--', label='Teoría')
    axs[i].set_title(f'Red de {n} nodos', fontsize=12)
    axs[i].set_xlabel('Número de experimentos', fontsize=10)
    axs[i].set_ylabel('Saltos promedio', fontsize=10)
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
